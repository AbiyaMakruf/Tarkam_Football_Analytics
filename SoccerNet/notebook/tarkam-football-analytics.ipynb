## Training Dataset SoccerNet

```python
# Download Dataset Dari Google Drive
!gdown 19sjiiEsyWK6lfGDx79lMDpXlb3wcvdFb

# Unzip Dataset
!unzip 5000.zip -d dataset_soccernet/
```

```python
# Train YOLOV11N
model_yolo11n = YOLO('yolo11n.pt')

results_yolo11n = model_yolo11n.train(
    data='dataset_soccernet/data.yaml',
    epochs=20,
    imgsz=640,
    device=[0,1] # Agar Bisa Menggunakan 2 GPU di Kaggle
)
```

```python
# Evaluasi Model Menggunakan Best Weight Untuk Subset Test
model_yolo11n.val(data='dataset_soccernet/data.yaml',split='test')
```

```python
# Train YOLOV11S
model_yolo11s = YOLO('yolo11s.pt')

results_yolo11s = model_yolo11s.train(
    data='dataset_soccernet/data.yaml',
    epochs=20,
    imgsz=640,
    device=[0,1] # Agar Bisa Menggunakan 2 GPU di Kaggle
)
```

```python
# Evaluasi Model Menggunakan Best Weight Untuk Subset Test
model_yolo11s.val(data='dataset_soccernet/data.yaml',split='test')
```
