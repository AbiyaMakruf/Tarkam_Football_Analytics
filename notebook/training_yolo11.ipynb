```python
from ultralytics import YOLO

# Load model hasil training atau pretrained (ganti dengan path ke model kamu)
model = YOLO("best.pt")  # atau 'yolov8n.pt' jika default

# Prediksi ke video
model.predict(
    source="view1_cut1.mp4",       # path ke video
    save=True,                # simpan hasil sebagai video baru
    conf=0.5,                # confidence threshold
    imgsz=640,                # ukuran gambar (opsional)
    save_txt=False,           # set True jika ingin simpan hasil sebagai label YOLO
    save_crop=False           # set True jika ingin potong deteksi per objek
)
```
